# Inledning

I en tidigare lektion fick vi lära oss *integraltecknets*, $\int$, betydelse och hur det används.
$ \int_a^b f(x)\, dx$ kan tolkas som arean mellan $x$-axeln och kurvan $f(x)$, från $x=a$ till $x=b$. 

I denna lektion kommer vi att behandla ett annat sätt att beskriva arean, nämligen som en *Riemann-summa*. Här tänker man sig att arean under kurvan är den sammanlagda arean av många små rektanglar under kurvan.


image.png

Nedan ser du en kod som visualiserar begreppet Riemann-summa för en funktion $f(x)=x^2-8x+16$, med gränserna $a=0$ och $b=10$ och en rektangelbredd på $h=0.5$ (testa att ändra gränserna, rektangelbredden och funktionen).

Notera dock att koden **inte beräknar arean** utan bara ritar rektanglarna.

```
# Import av viktiga bibliotek, denna cell måste köras för att de andra ska fungera!

import numpy as np                # bibliotek med matematiska funktioner
import matplotlib.pyplot as plt   # bibliotek för att göra grafer
```

```
# Visualisering av Riemann-summa

def f(x):
  return x**2 - 8*x + 16          # Funktionen f(x) = x^2 - 8x + 16 (x^2 skrivs x**2 i Python)

a = 2                             # vänster gränsvärde (start)
b = 10                            # höger gränsvärde   (stopp)
h = 0.2                          # steglängd

x1 = np.arange(a, b, 0.01)                       # skapar lista x1 = [a, a+0.01, a+0.02, ..., b-0.01]
plt.plot(x1, f(x1))                              # ritar linjer mellan alla punkter (x,f(x))

x = np.arange(a, b, h)                           # skapar lista x = [a, a+h, a+2*h, ..., b-h]

plt.bar(x, f(x), h, align="edge", edgecolor= 'black', alpha=0.5) 
    # ritar staplar vid varje värde av x, med höjd f(x), och bredd h
    # (align="edge" gör så rektangeln har vänster kant vid x (annars placeras de med mittpunkt vid x))
    # (edgecolor='black' gör staplarnas kanter svarta och synliga)                                             
    # (alpha=0.5 gör dem lite genomskinliga)

plt.grid(color='grey')
plt.show()
```

Nedan följer ett program för att faktiskt **beräkna värdet** på arean A: 

```
#@title Dubbelklicka här för att se hur programmet kan se ut

# Definiera kurvan / funktionen f(x)
def f(x):                       
    return x**2 - 8 * x + 16              # f(x) = x^2-8x+16

# Följande kodsnutt beräknar arean från x=a till x=b.
# Genom att stega igenom x från a till b, med steglängd h,
# beräknas arean av rektanglarna rekt_area = höjd*bredd ( f(x)*h ) 
# och addera ihop dessa till en summerad area (sum_area)

sum_area = 0          # initiera den summerade arean till 0
a = 0                 # vänstra gränsvärde (start)
b = 8                 # högra gränsvärde (stop)
h = 0.1               # definiera steglängd / bredden på rektanglarna
err = 0.01            # ibland avrundar datorn tal fel. Detta hjälper till för att undvika fel i while-loopen
x = a                 # initiera variabeln som kommer gå från a upp till det givna värdet b
while x < b-err:      # gör detta så länge x är mindre än b
  #höjd = f(x)                     # höjd på aktuell rektangel vid x
  #bredd = h                       # rektangelns bredd
  rekt_area = f(x)*h               # beräkna rektangelns area
  sum_area = sum_area + rekt_area  # lägg till rektangelns area till den summerade arean 
  x = x + h                        # öka x med steglängden h

print("Arean blir:", sum_area)     # skriv ut den summerade arean (från x=a upp till x=b)
```

# Areafunktionen


I detta avsnitt kollar vi på en **funktion** $A(x)$ som beräknar arean från $0$ till $x$, där $x$ kan varieras.

Areafunktion_klar.png

```
# Definiera kurvan / funktionen f(x)
def f(x):                       
  return x**2-8*x+16               # f(x) = x^2-8x+16

# Funktion A(x) som beräknar arean under grafen till f, från 0 till x.
def A(x): 
  sum_area = 0                      # initiera den summerade arean till 0
  h = 0                             # initiera variabeln som kommer gå från 0 upp till det givna värdet x
  delta_h = 0.5                     # definiera steglängd / bredden på rektanglarna
  err = 0.00001
  while h < x - err:                # gör detta så länge h är mindre än x
    #höjd = f(h)                    # höjd på aktuell rektangel vid h=x
    #bredd = delta_h                # rektangelns bredd
    rekt_area = f(h)*delta_h        # beräkna rektangelns area
    sum_area = sum_area + rekt_area # lägg till rektangelns area till den summerade arean
    h = h + delta_h                 # öka h med steglängden delta_h
  return sum_area                   # returnera den summerade arean (från 0 upp till x)
```

Testa att beräkna arean:

```
# Arean från 0 till 8
print("Arean blir:", A(1))
```
